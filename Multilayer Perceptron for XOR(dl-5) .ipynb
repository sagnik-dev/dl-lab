```python
import numpy as np

def step(x):
    if x >= 0:
      return 1
    else:
      return 0
w11=1
w12=1
w21=-1
w22=-1
w31=1
w32=1
b1=0.5
b2=1.5
b3=1.5

def xor_network(x1, x2):

    h1 = step((w11 * x1) + (w12 * x2) - b1)

    h2 = step((w21 * x1) + (w22 * x2) + b2)

    y = step((w31 * h1) + (w32 * h2) - b3)

    return y

inputs = [(0,0), (0,1), (1,0), (1,1)]

for x1, x2 in inputs:
    print(f"Input: {x1}, {x2} -> Output: {xor_network(x1, x2)}")
```
