
## for weight and bais
### w += Error_rate*x
### b += error_rate
### Here, Error_rate = target - output



```python
import numpy as np
# for AND gate
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 8
w,b = np.zeros(2),0


for _ in range(epochs):
  for x,y in zip(x_list,y_list):
    # print('y og :',y,",",end=" ")
    y_p = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if y_p >= 0 else 0
    er = y - y_pred
    print(er,"<- at epo",_)
    w += er*x
    b += er
    print(w,b)


```

```python
import numpy as np
# for AND gate
x_list = np.array([[2,2],[1,-2],[-2,2],[-1,1]])
y_list = np.array([0,1,0,1])
epochs = 3
w,b = np.zeros(2),0


for _ in range(epochs):
  for x,y in zip(x_list,y_list):
    # print('y og :',y,",",end=" ")
    y_p = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if y_p >= 0 else 0
    er = y - y_pred
    print("error rate : ",er,"<- at epo",_)
    w += er*x
    b += er
    print(w,b)
```
