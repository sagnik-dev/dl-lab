```python
import numpy as np
# for AND gate (Input/Output)
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 4
w,b = np.zeros(2),0

for _ in range(epochs):
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    print("error : ",E,"<- at epoch",_)
    w += E*x
    b += E
    print(w,b)
```

```python
#add logic 3rd
import numpy as np

# AND gate data
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])

epochs = 10

# random initialization
w,b = np.zeros(2),0


for epoch in range(epochs):
    all_zero_error = True   # assume perfect at start of epoch
    print("\nEpoch:", epoch)

    for x, y in zip(x_list, y_list):
        z = np.dot(x, w) + b
        y_pred = 1 if z > 0 else 0

        E = y - y_pred
        print("error:", E,"<- at epoch",_)

        if E != 0:
            all_zero_error = False

        w += E * x
        b += E

    # ðŸ”¹ Early stopping condition
    if all_zero_error:
        print("Training stopped: zero error for all inputs")
        break
```
